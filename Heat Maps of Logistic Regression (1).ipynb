```python
#lOGISTIC REGRESSION USING SQUUEZENET DATASET

import pandas as pd
# Load the CSV file
df = pd.read_csv('LogisticRegression CM.csv',header=None)  
# Convert the DataFrame to a NumPy array (matrix)
matrix = df.values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Class labels
class_labels = ["Arive-Dantu", "Basale", "Mint", "Pomegranate", "Drumstick", 
                "Karanda", "Jasmine", "Guava", "Peepal Tree", "Oleander", 
                "Mango", "Indian Beech", "Crape Jasmine", "Jackfruit", 
                "Jamaica Cherry-Gasagase", "Tulsi", "Rasna", "Roxburgh fig", 
                "Betel", "Mexican Mint", "Hibiscus Rosa-sinensis", "Parijata", 
                "Jamun", "Fenugreek", "Indian Mustard","Rose Apple","Sandalwood","Curry","Mango","Neem"]

# Generating a random 25x25 matrix (since there are 25 class labels)
#data = np.random.rand(30, 30)

# Create a heatmap
plt.figure(figsize=(15, 15))  # Adjust the size as needed
sns.heatmap(matrix, xticklabels=class_labels, yticklabels=class_labels, annot=False, cmap='viridis', cbar=True)

# Rotate labels for better visibility
plt.xticks(rotation=90,fontsize=24)
plt.yticks(rotation=0,fontsize=24)

#plt.title('Class Labels Heatmap')
plt.savefig('HeatMap on Squeezenet.jpeg', dpi=600, bbox_inches='tight')
plt.show()
```
