```python
import matplotlib.pyplot as plt
import pandas as pd
mcc_df = pd.read_csv('Evaluation Results.csv')
print(mcc_df)
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Assuming mcc_df has columns: 'Model', 'Accuracy', 'MCC'
# If not, replace the column names accordingly

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 16

# Assuming mcc_df has columns: 'Model', 'Accuracy', 'MCC'
# If not, replace the column names accordingly

mcc_df.plot(x='Model', y=['CA', 'MCC'], kind='bar', figsize=(12,8), stacked=False, width=0.8)

plt.xlabel(" ")
plt.ylabel("Accuracy")

plt.xticks(rotation=0, fontsize=16)
plt.yticks(fontsize=16)
#plt.legend(loc=(0.801, 0.007), fontsize=20)

plt.grid(axis='y')
plt.savefig('Evaluation Graph.jpeg', dpi=600, bbox_inches='tight')
plt.show()
```
